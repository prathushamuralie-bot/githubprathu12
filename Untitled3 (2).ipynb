Data cleaning and preparation

6.Handle missing values

```python
df = df.fillna(0)
```

7.check and remove duplicates

```python
print("Duplicate rows:",df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

8.convert date column into proper datetime format

```python
df['Date'] = pd.to_datetime(df['Date'])
```

9.Create a new column:Active Cases = Confirmed - (Recovered + Deaths)

```python
df['Active Cases'] = df['Confirmed']-(df['Recovered'] + df['Deaths'])
```

10.extract useful time-based features(year,month,day)

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
print(df.head())
```

Data Analysis

11.find total confirmed cases by country/region

```python
country_cases = df.groupby('Region')['Confirmed'].sum()
print(country_cases)
```

12.Identify top 5 countries with highest cases

```python
top5 = country_cases.sort_values(ascending=False).head(5)
print("Top 5 Countries:\n",top5)
```

13.Calculate total deaths and recovery rate

```python
total_deaths = df['Deaths'].sum()
total_confirmed = df['Confirmed'].sum()
total_recovered = df['Recovered'].sum()
recovery_rate = (total_recovered / total_confirmed) * 100
```

```python
print("total deaths:", total_deaths)
print("Recovery Rate:", recovery_rate)
```

14.Analyze daily or monthly growth in cases

```python
df  = df.sort_values('Date')
df['Daily Growth'] = df['Confirmed'].diff()
print(df[['Date', 'Daily Growth']].head())
```

```python
monthly = df.groupby(['Year', 'Month'])['Confirmed'].sum()
print(monthly)
```

15.Identify peak periods of infection

```python
peak_day = df.loc[df['Confirmed'].idxmax()]
print("peak day:\n", peak_day)
```

16.Compare confirmed vs recovered cases over time

```python
comparison = df.groupby('Date')[['Confirmed', 'Recovered']].sum()
print(comparison.head())
```
